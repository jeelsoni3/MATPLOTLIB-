```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for player performance
players = ['Ronaldo', 'Messi', 'Remos', 'Neymar', 'Christian']
goals_scored = [5, 8, 3, 7, 4]
assists = [3, 5, 2, 6, 3]
yellow_cards = [2, 1, 0, 3, 1]

# Create a player performance chart
fig, ax = plt.subplots()
bar_width = 0.3

index = np.arange(len(players))
bar1 = ax.bar(index, goals_scored, bar_width, label='Goals Scored')
bar2 = ax.bar(index + bar_width, assists, bar_width, label='Assists')
bar3 = ax.bar(index + 2 * bar_width, yellow_cards, bar_width, label='Yellow Cards')

ax.set_xlabel('Players')
ax.set_ylabel('Performance')
ax.set_title('Football Player Performance')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(players)
ax.legend()

# Show the player performance chart
plt.show()
```
